```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
games_df = pd.read_csv('nba_game_details_cleaned.csv')
games_df.head(10)
```

```python
games_df.TEAM_home.value_counts()
```

```python
games_df.WIN_home.value_counts()
```

```python
games_df.isna().sum()
```

```python
games_df = games_df.dropna()

games_df['TEAM_home'] = np.where(games_df['TEAM_home'] == 'NOH', 'NOP', games_df['TEAM_home'])
games_df['TEAM_away'] = np.where(games_df['TEAM_away'] == 'NOH', 'NOP', games_df['TEAM_away'])

games_df.head()
```

```python
games_df.TEAM_away.value_counts()
```

```python
df_corr = games_df.corr()
df_corr
```

```python
df_corr = games_df.corr()

# Drop fields
#'GAME_DATE', 'TEAM_away', 'TEAM_home',
fields = ['GAME_ID', 'TEAM_ID_away', 'TEAM_ID_home']
df_corr.drop(fields, inplace=True)
df_corr.drop(fields, axis=1, inplace=True)


df_corr
```

```python
np.ones_like(df_corr, dtype=np.bool)
```

```python
mask = np.triu(np.ones_like(df_corr, dtype=np.bool))
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
# mask
mask = np.triu(np.ones_like(df_corr, dtype=np.bool))
# adjust mask and df
mask = mask[1:, :-1]
corr = df_corr.iloc[1:,:-1].copy()
# plot heatmap
sb.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap='Blues',
           vmin=-1, vmax=1, cbar_kws={"shrink": .8})
# yticks
plt.yticks(rotation=0)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(24,20))

# mask
mask = np.triu(np.ones_like(df_corr, dtype=np.bool))
# adjust mask and df
mask = mask[1:, :-1]
corr = df_corr.iloc[1:,:-1].copy()
# color map
cmap = sb.diverging_palette(0, 230, 90, 60, as_cmap=True)
# plot heatmap
sb.heatmap(corr, mask=mask, annot=True, fmt=".2f", 
           linewidths=5, cmap=cmap, vmin=-1, vmax=1, 
           cbar_kws={"shrink": .8}, square=True)
# ticks
yticks = [i.upper() for i in corr.index]
xticks = [i.upper() for i in corr.columns]
plt.yticks(plt.yticks()[0], labels=yticks, rotation=0)
plt.xticks(plt.xticks()[0], labels=xticks)
# title and background
fig.patch.set_facecolor('white')
title = 'CORRELATION MATRIX\nNBA Team Stats\n'
plt.title(title, loc='left', fontsize=18)
plt.savefig('corr_matrix.png', facecolor=fig.get_facecolor(), edgecolor='none', dpi=100, transparent=False)
```
