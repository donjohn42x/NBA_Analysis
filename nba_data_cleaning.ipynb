## NBA game details data cleaning

1. Combine 10 years of game details
2. Filter neccesary columns
3. Create Game Number and WIN_home column
4. Combine every two rows into one row (Same GAME_ID)

```python
import pandas as pd
import os
```

```python
os.listdir("data/game_details")
```

```python
df_2012 = pd.read_csv("data/game_details/nba_game_detailsSeason=2012-13.csv")
df_2013 = pd.read_csv("data/game_details/nba_game_detailsSeason=2013-14.csv")
df_2014 = pd.read_csv("data/game_details/nba_game_detailsSeason=2014-15.csv")
df_2015 = pd.read_csv("data/game_details/nba_game_detailsSeason=2015-16.csv")
df_2016 = pd.read_csv("data/game_details/nba_game_detailsSeason=2016-17.csv")
df_2017 = pd.read_csv("data/game_details/nba_game_detailsSeason=2017-18.csv")
df_2018 = pd.read_csv("data/game_details/nba_game_detailsSeason=2018-19.csv")
df_2019 = pd.read_csv("data/game_details/nba_game_detailsSeason=2019-20.csv")
df_2020 = pd.read_csv("data/game_details/nba_game_detailsSeason=2020-21.csv")
df_2021 = pd.read_csv("data/game_details/nba_game_detailsSeason=2021-22.csv")
```

```python
df_list = [df_2012, df_2013, df_2014, df_2015, df_2016, df_2017,df_2018, df_2019, df_2020, df_2021]
```

```python
df_combined = pd.concat(df_list)
```

```python
df_combined.head()
```

```python
df_combined_filtered = df_combined.filter(['GAME_ID','GAME_DATE','MATCHUP','TEAM_ABBREVIATION','TEAM_ID','WL','FGM','FGA','FG_PCT','FG_3M','FG_3A','FG3_PCT','FTM','FTA','FT_PCT','OREB','DREB','REB','AST','STL','BLK','TOV','PF','PTS'])
df_combined_filtered = df_combined_filtered.rename(columns={'TEAM_ABBREVIATION':'TEAM'})
df_combined_filtered.head(10)
```

```python
home_away_list = []

for x in df_combined_filtered['MATCHUP']:
    if '@' in x:
        x = 'away'
        home_away_list.append(x)
    else:
        x = 'home'
        home_away_list.append(x)

df_combined_filtered['HOME_AWAY'] = home_away_list
df_combined_filtered.head()
```

```python
# Find the game which has no data
df_combined_filtered[df_combined_filtered['WL'].isna()]
```

```python
# Deleting the specific game from the dataframe
df_combined_filtered = df_combined_filtered.drop(df_combined_filtered[df_combined_filtered['GAME_ID']==21201214].index)
```

```python
df_combined_filtered.isna().sum()
```

```python
game_num = []
for i in range(0, int(len(df_combined_filtered)/2)):
    game_num.append(i)
    game_num.append(i)

df_combined_filtered['GAME_NUM'] = game_num
```

```python
# Convert two rows into one row / Clean up each column
df_new = df_combined_filtered.set_index(["GAME_NUM","HOME_AWAY"]).unstack()
df_new.columns = df_new.columns.map("_".join)
df_new.head(10)
```

```python
results = []

for result in df_new['WL_home']:
    if result == 'W':
        result = 1
        results.append(result)
    else:
        result = 0
        results.append(result)

df_new['WIN_home'] = results
```

```python
df_new = df_new.rename(columns={"GAME_DATE_away":"GAME_DATE","GAME_ID_away":"GAME_ID"})
df_new = df_new.drop(['GAME_ID_home',"GAME_DATE_home","MATCHUP_home","MATCHUP_away"], axis=1)
df_new['GAME_DATE'] = pd.to_datetime(df_new['GAME_DATE']) 
```

```python
df_new.head(20)
```

```python
df_new.to_csv('data/game_details/nba_game_details_cleaned.csv', index=False)
```
