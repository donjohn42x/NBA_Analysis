```python
import pandas as pd
import numpy as np
```

```python
games_df = pd.concat(map
                     (pd.read_csv, 
                          ['nba_game_detailsSeason=2021-22.csv', 
                           'nba_game_detailsSeason=2020-21.csv', 
                           'nba_game_detailsSeason=2019-20.csv',
                           'nba_game_detailsSeason=2018-19.csv',
                           'nba_game_detailsSeason=2017-18.csv',
                           'nba_game_detailsSeason=2016-17.csv',
                           'nba_game_detailsSeason=2015-16.csv',
                           'nba_game_detailsSeason=2014-15.csv',
                           'nba_game_detailsSeason=2013-14.csv',
                           'nba_game_detailsSeason=2012-13.csv']), ignore_index=True)
games_df.head(10)
```
